```python
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
dfs = pd.read_html(url)
```

```python
df = dfs[0]

df2 = df[['Postal Code', 'Borough', 'Neighbourhood']]
print(df2)
df2.head()
```

```python
data = 'http://cocl.us/Geospatial_data'
df3 = pd.read_csv(data)
df3.sort_values(by='Postal Code',ascending=True)
df3.head()
```

```python
Location = pd.merge(df2, df3)
Location
```
